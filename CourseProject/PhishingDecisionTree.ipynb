```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Load data
#phish_df = pd.read_csv('/content/phishingData.csv')
phish_df = pd.read_csv('phishData.csv')
phish_df = phish_df.dropna()

# Looking into all the column names
phish_df.columns
```

```python
# Creating decision tree only using the first 3 column
feature_1 = ['having_IP_Address', 'URL_Length', 'Shortining_Service']
target = ['Result']
X_1 = phish_df[feature_1]
y = phish_df['Result']
```

```python
clf = DecisionTreeClassifier() #clf to denote classifier
clf.fit(X_1,y)
```

```python
tree.plot_tree(clf, feature_names=feature_1)
```

```python
y_pred = clf.predict(X_1)
print(confusion_matrix(y, y_pred))
print(classification_report(y, y_pred))
```

As you can see when we are only using 3 attributes to predict whether or not the site was a phishing website, we have really poor accuracy as well really poor score or precision and recall as well.

```python
# adding more for data
feature_2 = ['having_IP_Address', 'URL_Length', 'Shortining_Service',
             'having_At_Symbol', 'double_slash_redirecting', 'Prefix_Suffix']
target = ['Result']
X_2 = phish_df[feature_2]
y = phish_df['Result']

clf = DecisionTreeClassifier() #clf to denote classifier
clf.fit(X_2,y)

tree.plot_tree(clf, feature_names=feature_2)
```

```python
y_pred = clf.predict(X_2)
print(confusion_matrix(y, y_pred))
print(classification_report(y, y_pred))
```

As we can see giving adding more data to the algorithm is giving us a slightly improved rusult.
